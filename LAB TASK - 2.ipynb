```python
# Initial movie dataset
movies = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Memento", 9000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]

# Optional: Add user input movies
add_more = input("Do you want to add more movies? (yes/no): ").strip().lower()

if add_more == "yes":
    try:
        count = int(input("How many movies do you want to add? "))
        for _ in range(count):
            name = input("Enter movie name: ")
            budget = int(input("Enter movie budget: "))
            movies.append((name, budget))
    except ValueError:
        print("Invalid input. Skipping adding movies.\n")

# 1. Calculate average budget
total_budget = sum(budget for _, budget in movies)
average_budget = total_budget / len(movies)

print(f"\nðŸŽ¥ Total movies: {len(movies)}")
print(f"ðŸ’° Average movie budget: ${average_budget:,.0f}\n")

# 2. Find and print high budget movies
high_budget_count = 0

print("ðŸ” Movies with budget higher than average:\n")
for name, budget in movies:
    if budget > average_budget:
        difference = budget - average_budget
        high_budget_count += 1
        print(f"ðŸŽ¬ {name} â†’ ${budget:,.0f} (Higher by ${difference:,.0f})")

# 3. Count high budget movies
print(f"\nðŸ“Š Number of movies above average budget: {high_budget_count}")
```
